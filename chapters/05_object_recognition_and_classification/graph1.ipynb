```python
import math

def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)
plt.plot(x,sig)
plt.show()
```

```python
x = np.arange(-10., 10., 0.2)
plt.plot(x, np.tanh(x))
plt.show()
```

```python
x = np.arange(-5., 5., 0.2)
plt.plot(x, np.maximum(x,0))
plt.show()
```
